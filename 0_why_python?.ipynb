```
author = "Guido van Rossum"
date = 1989
foo = "%s started Python in %d" % (author, date)
print(foo)
```

```
foo = "{} started Python in {}".format(author,date)
print(foo)
```

```
foo = "{author} started Python in {date}".format(author=author,date=date)
print(foo)
```

### This can only be done in Python 3+

```
foo = f"{author} started Python in {date}"
print(foo)
```
